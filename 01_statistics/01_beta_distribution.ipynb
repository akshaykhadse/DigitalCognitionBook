# Beta Distribution

## Load Libraries

```
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
```

Lets define PDF of Beta distribution.
$$f(x \mid \alpha, \beta) = \frac{x^{\alpha-1}(1-x)^{\beta-1}}{\mathcal{B}(\alpha, \beta)}$$

Here, $\mathcal{B}(\alpha, \beta)$ is called the Beta function and it is given by

$$
\begin{align}
\mathcal{B}(\alpha,\beta) &= \int_{0}^{1} x^{\alpha-1}(1-p)^{\beta -1} dx\\
&= \frac{\Gamma(\alpha) \Gamma(\beta)}{\Gamma(\alpha + \beta)}
\end{align}
$$

where, $\Gamma(n) = (n-1)!$ for all $n \in \mathbb{Z}^{+}$.

```
from math import factorial
def beta(x, a, b):
    mu=0
    sigma=1
    y = ss.beta.pdf(x, a, b, mu, sigma)
    return y
```

Let's plot the beta distribution for some values of $\alpha$ and $\beta$

```
x = np.linspace(0, 1, 100) # Domain of Beta distribution

plt.figure(figsize=(8,8))

# alpha = beta = 0.5
a1 = 0.5
b1 = 0.5
y1 = beta(x, a1, b1)
plt.plot(x, y1)

# alpha = 5, beta = 1
a2 = 5
b2 = 1
y2 = beta(x, a2, b2)
plt.plot(x, y2)

# alpha = 1, beta = 3
a3 = 1
b3 = 3
y3 = beta(x, a3, b3)
plt.plot(x, y3)

# alpha = 2, beta = 2
a4 = 2
b4 = 2
y4 = beta(x, a4, b4)
plt.plot(x, y4)

# alpha = 2, beta = 5
a5 = 2
b5 = 5
y5 = beta(x, a5, b5)
plt.plot(x, y5)

# Legend
plt.legend(["alpha = beta = 0.5", \
            "alpha = 5, beta = 1", \
            "alpha = 1, beta = 3", \
            "alpha = 2, beta = 2", \
            "alpha = 2, beta = 5"])

plt.xlim(0, 1)
plt.ylim(0, 2.5)
plt.grid(ls = ':');
```

Let's plot the distribution for $\alpha=10$ and $\beta=10$.

```
x = np.linspace(0, 1, 100) # Domain of Beta distribution

plt.figure(figsize=(8,6))

# alpha = beta = 0.5
a = 10
b = 10
y = beta(x, a, b)
plt.plot(x, y)

plt.xlim(0, 1)
plt.ylim(0, 3.6)
plt.grid(ls = ':');
```

Observe that this mound shaped distribution looks similar to normal distribution except two differences:
- The peak is at $x=0.5$
- The value of PDF $f(x = 0.5 \mid 10, 10) = 3.5$
