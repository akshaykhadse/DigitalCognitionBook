# Linear Regression
This example uses the [Advertising Dataset](https://www.kaggle.com/ashydv/advertising-dataset) available on Kaggle.

## Setup Advertising Dataset
Download the advertising.csv file from git repo. The original data can be found at [Advertising Dataset](https://www.kaggle.com/ashydv/advertising-dataset) Kaggle page.

## Load libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

## Read the data


```
data = pd.read_csv("advertising.csv")
data.head()
```

## Modelling
Let's visualize the effect of TV commercials on the sales

```
plt.figure(figsize=(8, 6))
plt.scatter(
    data['TV'],
    data['Sales'],
    marker="o",
    edgecolor='black',
    facecolors='none'
)
plt.xlabel("TV ($)");
plt.ylabel("Sales ($)");
```

There is a clear relationship between the amount spent on TV ads and sales.

Let's fit a linear regression model to this data.

```
X = data['TV'].values.reshape(-1,1)
y = data['Sales'].values.reshape(-1,1)
reg = LinearRegression()
reg.fit(X, y)
print("The linear model is: Y = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```

Let’s visualize how the line fits the data

```
predictions = reg.predict(X)
plt.figure(figsize=(8, 6))
plt.scatter(
    data['TV'],
    data['Sales'],
    marker="o",
    edgecolor='black',
    facecolors='none'
)
plt.plot(
    data['TV'],
    predictions,
    c='blue',
    linewidth=2
)
plt.xlabel("TV ($)");
plt.ylabel("Sales ($)");
```

It seems that a simple linear regression can explain the general impact of amount spent on TV commercial and sales
